```
###Download From source into GDrive using Colab

import requests  
file_url = "https://www.kaggle.com/grassknoted/asl-alphabet/download"
    
r = requests.get(file_url, stream = True)  
  
with open("/content/gdrive/My Drive/python.pdf", "wb") as file:  
    for block in r.iter_content(chunk_size = 1024): 
         if block:  
             file.write(block)  
```
